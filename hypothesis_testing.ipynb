```python
import pandas as pd
from scipy.stats import ttest_ind, mannwhitneyu
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV file
data = pd.read_csv('data_to_use/average_time_per_video.csv')

# Separate data into in-cycle and out-of-cycle groups
in_cycle = data[data['In Cycle'] == 'Yes']
out_cycle = data[data['In Cycle'] == 'No']

# Analyze 'Watch Hours'
in_cycle_hours = in_cycle['Watch Hours']
out_cycle_hours = out_cycle['Watch Hours']

# Perform t-test
t_stat, t_p_value = ttest_ind(in_cycle_hours, out_cycle_hours, equal_var=False)

# Perform Mann-Whitney U test (non-parametric)
u_stat, u_p_value = mannwhitneyu(in_cycle_hours, out_cycle_hours, alternative='two-sided')

# Print results
print("T-Test Results:")
print(f"T-Statistic: {t_stat}, P-Value: {t_p_value}")
print("\nMann-Whitney U Test Results:")
print(f"U-Statistic: {u_stat}, P-Value: {u_p_value}")

# Determine significance
if t_p_value < 0.05:
    print("Significant difference found using T-Test.")
else:
    print("No significant difference found using T-Test.")

if u_p_value < 0.05:
    print("Significant difference found using Mann-Whitney U Test.")
else:
    print("No significant difference found using Mann-Whitney U Test.")

# Visualization: Box Plot
# Create a new column for group labels
data['Cycle Status'] = data['In Cycle'].replace({'Yes': 'In Cycle', 'No': 'Out of Cycle'})

# Plot the boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x='Cycle Status', y='Watch Hours', data=data, palette='pastel')
plt.title("Box Plot of Watch Hours During and Outside Cycle")
plt.xlabel("Cycle Status")
plt.ylabel("Watch Hours")
plt.show()
```
